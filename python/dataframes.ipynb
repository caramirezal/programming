```python
## get null values
data.isnull().head()
```

```python
## add new column
data["newCol"] = np.array([5]*len(data))
data.head()
```

```python
## drop any nan values
data.dropna(how="any").head()
```

```python
## fill nan values with zeros
data.fillna(0)
```

```python
## calculate column means
data.mean()[:10]
```

```python
# calculate row means 
data.mean(1)[:10]
```

```python
## operation between columns
data["multCol"] = data.apply(lambda col: col["Ubc"] * col["newCol"], axis=1)
data.head()
```

```python
## Another form to operate between columns
data["multCol"] = data["Ubc"] * data["newCol"]
data.head()
```
