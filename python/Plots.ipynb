```python
from datascience import *
import numpy as np

## Import plot functions
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
```

```python
data = Table.read_table("/home/carlos/scripts/dataScience/data/NIHMS53736-supplement-9.csv")
data.show(3)
```

```python
## selecting one column to plot
data_norm = data.select("Cell","Abl1")
## adding an index column for x-axis
data_norm = data_norm.with_column(
    'Cell', np.arange(data_norm.num_rows)
)

## plotting only the first 30 values
data_norm.take(np.arange(30)).plot("Cell")
```

```python
## scatter plot
data_norm.take(np.arange(30)).scatter(0,1)
```

```python
## bar histogram
data.take(np.arange(10)).barh("Cell","Abl1")
```
